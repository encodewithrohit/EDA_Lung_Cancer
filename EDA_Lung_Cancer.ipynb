```python
# To further investigate the relationship between lung cancer diagnosis and other variables, let's create some visualizations and perform additional statistical analyses:
'''
- Distribution of Lung Cancer by Gender
- Age Distribution in Lung Cancer Patients
- Bar plots for categorical variables against Lung Cancer diagnosis
- Chi-square tests for independence to see if there is a statistically significant relationship between each categorical variable and lung cancer diagnosis
'''
```

```python
# Start with doing vizualization imports
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency

# Setup the figure size
plt.figure(figsize=(12,8))

# Distribution of Lung Cancer by Gender
plt.subplot(2,2,1)
sns.countplot(data=dataset, x="GENDER", hue="LUNG_CANCER")
plt.title("Lung cancer ditribution by Gender")

# Age distribution in Lung Cancer patients
plt.subplot(2,2,2)
sns.histplot(dataset[dataset["LUNG_CANCER"] == "YES"]["AGE"], kde=True, color="red", label="Lung Cancer")
sns.histplot(dataset[dataset["LUNG_CANCER"] == "NO"]["AGE"], kde=True, color="blue", label="No Lung Cancer")
plt.title("Age distribution")
plt.legend()

# Bar plots for categorical variables against Lung Cancer diagnosis
categorical_vars = ['SMOKING', 'YELLOW_FINGERS', 'ANXIETY', 'PEER_PRESSURE',
                    'CHRONIC_DISEASE', 'FATIGUE', 'ALLERGY', 'WHEEZING',
                    'ALCOHOL_CONSUMING', 'COUGHING', 'SHORTNESS_OF_BREATH',
                    'SWALLOWING_DIFFICULTY', 'CHEST_PAIN']

plt.figure(figsize=(20,20))
for i, var in enumerate(categorical_vars, 1) :
  plt.subplot(4, 4, i)
  sns.countplot(data=dataset, x=var, hue='LUNG_CANCER')
  plt.title(f'Lung Cancer by {var}')

plt.tight_layout()
plt.show()

# Chi-square tests for independence
chi2_results = {}
for var in categorical_vars :
  contingency_table = pd.crosstab(dataset[var], dataset['LUNG_CANCER'])
  chi2, p, _, _ = chi2_contingency(contingency_table)
  chi2_results[var] = p

chi2_results
```
